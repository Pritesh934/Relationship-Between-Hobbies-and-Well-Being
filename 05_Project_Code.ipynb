```python
import numpy as np

heights=df2['height_in_inches'].astype(str).str.strip()

pattern = r"(\d+)['\.]\s*(\d*)"
extracted = heights.str.extract(pattern)

feet = pd.to_numeric(extracted[0], errors='coerce')
inches = pd.to_numeric(extracted[1], errors='coerce').fillna(0)

height_ft_in_np = (feet * 12) + inches

height_ft_in_np = np.where(feet < 8, height_ft_in_np, np.nan)
height_ft_in = pd.Series(height_ft_in_np, index=df2.index)

raw_numbers = pd.to_numeric(heights.str.replace(r"[^\d.]", "", regex=True), errors='coerce')

final_height = height_ft_in.combine_first(raw_numbers)

final_height = np.where(final_height > 100, final_height / 2.54, final_height)

df2['height_in_inches']=final_height
df2
```

**1.** Do students with active hobbies, such as arts, music, or sports, have better GPAs than students with no hobbies? **(Linear Regression)**

```python
import statsmodels.api as sm

print("Unique values in Hobbies column:", df2['hobbies'].unique())
no_hobby_indicators = ['None', 'none', 'Nothing', 'nothing', 'No', 'no', np.nan]

df2['has_active_hobby'] = df2['hobbies'].apply(lambda x: 0 if x in no_hobby_indicators else 1)
```

```python
Y = df2['gpa']
X = df2['has_active_hobby']

X = sm.add_constant(X)
model_q1 = sm.OLS(Y, X).fit()

print(model_q1.summary())
```

**Interpretation:**

- Since the coefficient for **has_active_hobby** is 3.3451 (+ve), we can say that students with active hobbies have a higher GPA on average. Having a hobby is associated with a 3.3451 points increase in GPA.

- Additionally, the P-Value for **has_active_hobby** is 0.910 which is larger than 0.05. This suggests that the relationship between GPA of students and them having active hobbies is not statistically significant.


- The Constant (const) basically represents the average GPA of students with no hobbbies as X = 0 for them.

- We cannot definitely conculde that people who have hobbies do not necessarily have higher GPAs than people who do not have hobbies based on our data.

- From our analysis we find that the relationship between GPA and having active hobbies is not statistically significant, which implies that we cannot say that those who have hobbies necessarily have a higher GPA than those who do not have any hobbies.

**3.** Are there significant patterns or clusters among students’ preferences (e.g., beverage/cuisine, media/hobbies, sleep patterns) related to academic or social outcomes? **(K-Means Clustering)**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder


features = [
    'average_hours_of_sleep_per_night',
    'hours_spent_on_social_media_per_week',
    'bevarage_preference',
    'favorite_cuisine',
    'night_owl_or_early_bird',
    'has_active_hobby'
]


numeric_cols = ['average_hours_of_sleep_per_night', 'hours_spent_on_social_media_per_week', 'has_active_hobby']
categorical_cols = ['bevarage_preference', 'favorite_cuisine', 'night_owl_or_early_bird']


for col in numeric_cols:
    if df2[col].dtype == 'object':
        df2[col] = df2[col].astype(str).str.replace(r'[^0-9.]', '', regex=True)
        df2[col] = pd.to_numeric(df2[col], errors='coerce')
        df2[col].fillna(df2[col].mean(), inplace=True)


df2['average_hours_of_sleep_per_night'] = df2['average_hours_of_sleep_per_night'].clip(lower=2, upper=14)


X_cat = pd.get_dummies(df2[categorical_cols], drop_first=True)

scaler = StandardScaler()
X_num = pd.DataFrame(scaler.fit_transform(df2[numeric_cols]), columns=numeric_cols, index=df2.index)

X_clustering = pd.concat([X_num, X_cat], axis=1)
```

```python
from sklearn.cluster import KMeans

inertia = []
K_range = range(1, 11)

print("Calculating elbow curve...")

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    kmeans.fit(X_clustering)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Total Distance)')
plt.title('Elbow Method (Smoothed with n_init=20)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

```python
k = 4
kmeans_final = KMeans(n_clusters=k, random_state=42, n_init=10)
df2['cluster_label'] = kmeans_final.fit_predict(X_clustering)

print(f"Cluster Analysis (k={k}):\n")

analysis = df2.groupby('cluster_label')[['gpa', 'hours_spent_studying_per_week', 'average_hours_of_sleep_per_night']].mean()
print(analysis)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='average_hours_of_sleep_per_night', y='gpa', data=df2, hue='cluster_label')
plt.title('GPA Distribution by Preference Cluster')
plt.xlabel('average_hours_of_sleep_per_night')
plt.ylabel('GPA')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='hours_spent_studying_per_week', y='gpa', data=df2, hue='cluster_label')
plt.title('GPA Distribution by Preference Cluster')
plt.xlabel('average_hours_of_sleep_per_night')
plt.ylabel('GPA')
plt.show()
```

**Interpretation:** Academic success (GPA 3.65) is linked to a balance of study and sleep, while excessive studying (81.7 hours/week) does not increase GPA. The lowest-performing students (GPA 2.8) have similar study/sleep habits to high performers, suggesting non-lifestyle factors significantly impact academic outcomes.

**4.** Are students who identify as ‘extroverts’ more likely to report multiple hobbies compared to ‘introverts’? **(Classification: Logistic Regression)**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import statsmodels.api as sm

def count_hobbies(hobby_string):

    if pd.isna(hobby_string) or str(hobby_string).lower() in ['none', 'nothing', 'nan', '0']:
        return 0

    return len(str(hobby_string).split(','))

df2['hobby_count'] = df2['hobbies'].apply(count_hobbies)


print("Personality types found:", df2['introvert_or_extrovert'].unique())

df2['is_extrovert'] = df2['introvert_or_extrovert'].apply(
    lambda x: 1 if str(x).lower().strip() == 'extravert' else 0
)
```

```python
X = df2[['hobby_count']]
X = sm.add_constant(X)
Y = df2['is_extrovert']

logit_model = sm.Logit(Y, X).fit()
print(logit_model.summary())

X_train, X_test, y_train, y_test = train_test_split(df2[['hobby_count']], Y, test_size=0.3, random_state=42)

clf = LogisticRegression()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

plt.figure(figsize=(8, 6))
sns.barplot(x='introvert_or_extrovert', y='hobby_count', data=df2, palette='pastel', hue='introvert_or_extrovert')
plt.title('Average Number of Hobbies: Introverts vs. Extroverts')
plt.ylabel('Average Count of Hobbies')
plt.show()
```

**Interpretation:** Based on the Logistic Regression output, the results indicate no significant relationship between the number of hobbies a student has and whether they identify as an extrovert. The trend suggests that ambiverts have more hobbies than extroverts and introverts
